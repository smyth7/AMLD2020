```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math
```

```python
def gaussian(mu, sigma, x): return 100*stats.norm.pdf(x, mu, sigma)

def affine(a, b, x): return a*x + b

def sigmoid(x): return 1/(1+np.exp(-x))

def binary_cross_entropy(x): return -np.log(sigmoid(x))

def mse(x): return (1-x)**2
```

```python
def plot_fun(ax, x_min, x_max, fun, pts=100, fmt='-', **kwargs):
    x = np.linspace(x_min, x_max, pts)
    y = [fun(i) for i in x]
    if x_min is not None:
        x = [i for i in x if i > x_min]
        y = y[-len(x):]
    if x_max is not None:
        x = [i for i in x if i < x_max]
        y = y[:len(x)]
    ax.plot(x, y, fmt, **kwargs)
```

```python
x_min = -2
x_max = 4

plt.figure(figsize=(6, 4), dpi=200)
plt.xlabel("d = discriminator output for a fake sample")
ax1 = plt.gca()
c1 = 'orange'
bound = 0#-np.log(np.exp(1)-1)
plot_fun(ax1, x_min, x_max, binary_cross_entropy, color=c1)
ax1.set_ylabel("cross-entropy(sigmoid(d), target=1)", color=c1)
ax1.tick_params(axis='y', labelcolor=c1)

ax2 = ax1.twinx()
c2 = 'r'
plot_fun(ax2, x_min, x_max, mse, color=c2)
ax2.set_ylabel('MSE(d, target=1)', color=c2)
ax2.tick_params(axis='y', labelcolor=c2)

plt.title("Generator loss as a function of Discriminator output")
plt.show()
```

```python
real_data = lambda x: gaussian(2, 1, x)
fake_data = lambda x: gaussian(-2, 5, x)
```

```python
x_min = -10
x_max = 10
bound_real_fake = 2.5
bound_far_close = -2.4

plt.figure(figsize=(6, 4), dpi=200)
plt.xlabel("Sample value")
plt.ylabel("% of data")
ax1 = plt.gca()
ax2 = ax1.twinx()

plot_fun(ax1, x_min, x_max, real_data, color='g', label="Real data distribution")
plot_fun(ax1, x_min, x_max, fake_data, color='b', fmt='-', label="Fake data distribution")
ax1.legend(loc='upper center', bbox_to_anchor=(0.15, -0.15))

c = 'orange'
discriminator = lambda x: sigmoid(affine(4, 0, x))
plot_fun(ax2, x_min, 0.5, discriminator, pts=300, fmt='-', color=c, label="sigmoid(d) - high crossentropy gradient")
plot_fun(ax2, 0.5, x_max, discriminator, pts=300, fmt=':', color=c, label="sigmoid(d) - low crossentropy gradient")
ax2.set_ylabel('sigmoid(d)', color=c)
ax2.tick_params(axis='y', labelcolor=c)
ax2.legend(loc='upper center', bbox_to_anchor=(0.75, -0.15))

plt.show()
```
